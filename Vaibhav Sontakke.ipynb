## Task 4
Find out monthly sales and analyze findings

```python
#create separate dataset with two colounms for year-month and sales
ds['Month year']=ds['Order Date'].apply(lambda x:x.strftime('%Y-%m'))
monthly =pd.DataFrame( ds.groupby('Month year').sum()['Sales'].reset_index())
#print this dataset
monthly
```

```python
#plot this dataset
plt.figure(figsize=(15,6))
plt.plot(monthly['Month year'],monthly['Sales'])
plt.xticks(rotation='vertical', size=10)
plt.show()
```

From the plot of monthly sales we can see that sales are low for January and February then it starts increasing near March and April we can see small peak. After that sales are average until august when it starts increasing again. Sales are good from August to December except for  month of October. For months January ,February, October when sales are low we can give discount or other offers to attract people. we can aim for less profit in these months and try to compensate when sales are at its peak.Sales were maximum for November of 2017  

## Task 5
Find out yearly sales and analyze findings

```python
#create separate dataset with two colounms for year-month and sales
ds['Year']=ds['Order Date'].apply(lambda x:x.strftime('%Y'))
yearly =pd.DataFrame( ds.groupby('Year').sum()['Sales'].reset_index())
#print this dataset
yearly
```

```python
#plot this data
plt.figure(figsize=(15,6))
plt.plot(yearly['Year'],yearly['Sales'])
plt.xticks(rotation='vertical', size=10)
plt.show()
```

As we can see sales are low for first two years but after 2015 it increased from 500,000 to 700,000 and above. It is getting better every year. It shows that whatever the current strategy is ,it’s working and we can continue it. 

## Task 6
Finding monthly growth rate and analyse findings

```python
# create monthly growth rate coloum in monthly sales dataset
monthly['Growth Rate']=monthly['Sales'].pct_change()*100
monthly
```

```python
# maximum growth rate from data
monthly['Growth Rate'].max()
```

```python
plt.figure(figsize=(15,6))
plt.plot(monthly['Month year'],monthly['Growth Rate'])
plt.xticks(rotation='vertical', size=10)
plt.xlabel('month and year')
plt.ylabel('monthly growth rate in %')
plt.show()
```

Maximum growth rate was in march of 2014 which is 1132.131409 and minimum in January of 2016 which is -75.250121. Its interesting to note that growth rate was never as high as in march of 2014 ever again not even close. Closest it get is in march of 2015 which is 224.030794
